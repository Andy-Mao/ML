```python
import numpy as np
import pandas as pd
import os

from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset,TabularPredictor
```

```python
os.chdir("train") #更改路径
```

```python
base_info=pd.read_csv("base_info.csv")
entprise_info=pd.read_csv("entprise_info.csv")
```

```python
info=pd.merge(entprise_info,base_info,on='id',how='left')
```

```python
info=TabularDataset(info)  
info=info.drop(['id'],axis=1)
```

```python
train,test=train_test_split(info,random_state=0,test_size=0.3)
```

```python
predictor=TabularPredictor(label='label').fit(train,
#                                            num_stack_levels=1,
#                                             num_bag_folds=5,
                                              presets='best_quality')
```

```python
preds=predictor.predict(test.drop(columns=['label']))
```

```python
predictor.evaluate(test)
```

```python
predictor.leaderboard(train,silent=True)
```

```python
feature_importance = predictor.feature_importance(train)
feature_importance
```
